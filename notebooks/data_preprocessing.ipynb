### Insights from our raw data so far
- Addressline 2 should be removed due to large chunk of missing values and phone column is not relevant
- Explore Data using ORDERNUMBER
- Replace Null Values
- Some column data types need to be changed
- Change column names for better readability


```python
df.drop(columns=["ADDRESSLINE2", "PHONE"], inplace=True)
```

```python
pd.read_sql("SELECT * FROM sales_raw_data WHERE STATE IS NULL", conn)
```

```python
df["TERRITORY"].value_counts(dropna=False)
```

```python
df['ORDERNUMBER'].value_counts(ascending=False)
```

```python
df["POSTALCODE"] = df["POSTALCODE"].astype("string")
df["POSTALCODE"] = df["POSTALCODE"].fillna("UNKNOWN")
```

```python
df["STATE"] = df["STATE"].astype("string")
df["STATE"] = df["STATE"].fillna("UNKNOWN")
```

```python
df['TERRITORY'] = df['TERRITORY'].astype('string')
df['TERRITORY'] = df['TERRITORY'].fillna('UNKNOWN')
```

```python
# Change dtypes
df[df.select_dtypes(include='object').columns] = df.select_dtypes(include="object").astype("string")
df = df.astype({"ORDERDATE": "datetime64[ns]",
                "ORDERLINENUMBER": "string"})
```

```python
pd.read_sql("SELECT * FROM sales_raw_data WHERE ORDERNUMBER = 10332 LIMIT 5", conn)
```

```python
# Standardize column names
df.columns = (df.columns.str.strip().str.lower().str.replace(' ', '_'))
```

```python
df.columns
```

```python
# Change column names manually
df.rename(columns={"ordernumber": "order_no",
                   "quantityordered":"quantity",
                    "priceeach": "price_per_unit",
                    "orderlinenumber": "order_line_no",
                    "orderdate": "order_date",
                    "productline": "product_line",
                    "productcode": "product_code",
                    "customername": "customer_name",
                    "addressline1": "address",
                    "postalcode":"postal_code",
                    "contactlastname": "contact_last_name",
                    "contactfirstname": "contact_first_name",
                    "dealsize": "deal_size"}, inplace=True)
```

```python
df[["price_per_unit", "msrp"]]
```

```python
# Create new columns to calculate discount
df["discount_amount"] = df["msrp"] - df["price_per_unit"]
df["discount_pct"] = ((df["msrp"] - df["price_per_unit"]) / df["msrp"] * 100).round(2)
```

```python
df.to_csv("clean_data.csv", index=False)
```
